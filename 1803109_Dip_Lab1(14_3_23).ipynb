```python
import numpy as np   
# Image data can be represented as arrays of pixel values, where each pixel contains information about the color or brightness of a specific point in the image.
# Numpy provides an efficient way to work with arrays of numerical data or pixel value.
# NumPy provides tools for reading and writing image data in various formats, such as JPEG, PNG, and TIFF.
# NumPy makes it easy to perform operations on image data, such as cropping, resizing, and rotating images.
# NumPy provides tools for filtering and enhancing images, such as applying blur, sharpening, and contrast adjustments
# NumPy provides a wide range of mathematical operations that can be used in image processing, such as convolution and Fourier transforms.


import cv2
# cv2 is part of the OpenCV (Open Source Computer Vision) project, which is a collection of programming functions and algorithms that are commonly used in computer vision applications.
# cv2 provides tools for detecting and extracting features from images, such as corners, edges, and contours.
# some characteristics of np also include here.
import matplotlib.pyplot as plt
#Matplotlib is a popular Python library that is commonly used in digital image processing for data visualization.for example  various types of plots and charts, including line plots, scatter plots, bar graphs, and more.

from google.colab.patches import cv2_imshow
#The cv2_imgshow() function takes an image as input and displays it in a window on the screen. It can be used to quickly and easily visualize image data, which is useful in many digital image processing applications.
```

```python
image_gray.ndim # In digital image processing, the ndim attribute is used to get the number of dimensions of a NumPy array. For grayscale images, the image_gray variable would typically be a 2D NumPy array, representing the image as a matrix of pixel intensities.
```

```python
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) #here image convert from BGR to HSV .
# the HSV color space represents colors as combinations of hue, saturation, and value.
# The cv2.cvtColor() function is then used to convert the image from the BGR color space (which OpenCV uses by default) to the HSV color space. The resulting image is stored in the image_hsv variable as a NumPy array with three channels, representing the hue, saturation, and value components of the image.
# The hue ranges from 0 to 360 degrees, with red at 0 degrees, green at 120 degrees, and blue at 240 degrees.
# The saturation component represents the purity or intensity of the color, and ranges from 0 to 1. A saturation value of 0 represents grayscale, while a saturation value of 1 represents a fully saturated color.
# The value component represents the brightness or luminosity of the color, and ranges from 0 to 1. A value of 0 represents black, while a value of 1 represents white.
```

```python
cv2_imshow(image_hsv) #Display HSV image
```

```python
plt.imshow(image_hsv)  #here show HSV image using mathplotlib..
```

```python
# way of spliting red,green,blue channel...
b,g,r=cv2.split(image)
```

```python
b # array of pixel for blue channel
```

```python
g # array of pixel for green channel
```

```python
r # array of pixel for red channel
```

```python
rgb_image=cv2.merge((r,g,b)) # cv2.merge() function merge three channel(red,green,blue) into a single color image.
```

```python
#no of blue pixel..
num_of_blue_pixels=cv2.countNonZero(b)
print(num_of_blue_pixels)

# cv2.countNonZero() function from the OpenCV library is used to count the number of non-zero (i.e., non-black) pixels
```

```python
#no of green pixel..
num_of_green_pixels=cv2.countNonZero(g)
print(num_of_green_pixels)
```

```python
#no of red pixel..
num_of_red_pixels=cv2.countNonZero(r)
print(num_of_red_pixels)
```
